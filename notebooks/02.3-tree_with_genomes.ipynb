```python
# Filter data to only contain P5, P8, and P9
#df = pd.read_csv('../data/qiime2/feature-table.tsv', sep='\t', skiprows=1, index_col=0)
df = df.filter(regex=("P.*")) #select all that starts with P
df = df.filter(regex=('^(?!Pmix)')) #remove all that starts with Pmix
df = df.loc[~(df==0).all(axis=1)] #drop if row is all zeroes
# Split matrix per sampling sites
def per_sample(df, name):
    df = df.filter(regex=(name))
    df = df.loc[~(df==0).all(axis=1)]
    return df

df_split = []
for i in ['P5', 'P8', 'P9']:
    df_i = per_sample(df, i)
    df_split.append(df_i)
    
df_P5 = df_split[0]
df_P8 = df_split[1]
df_P9 = df_split[2]

for a in df_split:
    for i in a.columns:
        a.loc[:,i] = a.loc[:,i] / a.loc[:,i].values.sum() / 5
        
df_sum = []
for i in df_split:
    a = i.sum(axis=1).to_frame()
    df_sum.append(a)

#df_sum[1].sort_values(by=0, ascending=False)
df_sum[0] = df_sum[0].rename(columns={0:'P5'})
df_sum[1] = df_sum[1].rename(columns={0:'P8'})
df_sum[2] = df_sum[2].rename(columns={0:'P9'})
df_sum = pd.concat(df_sum).fillna(0)
```

```python
from ete3 import CircleFace
ts = TreeStyle()
root = anc

# Once you have your tree loaded, you need to associate matrix vectors to the tip nodes
for lf in root.iter_leaves():
    if lf.name in df_bin.index:
        if df_bin.loc[lf.name, 'P5'] == 1:
            lf.add_face(CircleFace(10, "Red"), column=0, position="aligned")
        if df_bin.loc[lf.name, 'P8'] == 1:
            lf.add_face(CircleFace(10, "Green"), column=1, position="aligned")        
        if df_bin.loc[lf.name, 'P9'] == 1:
            lf.add_face(CircleFace(10, "Blue"), column=2, position="aligned")
    else:
        pass
```

```python
ts = TreeStyle()

def mylayout(node):
    nstyle = NodeStyle()
    nstyle["fgcolor"] = "darkred"
    node.set_style(nstyle)
    if node.is_leaf() and node.name.startswith("P5"):
        nst1 = NodeStyle()
        nst1["bgcolor"] = "pink"
        node.set_style(nst1)
    elif node.is_leaf() and node.name.startswith("P8"):
        nst2 = NodeStyle()
        nst2["bgcolor"] = "lightgreen"
        node.set_style(nst2)
    elif node.is_leaf() and node.name.startswith("P9"):
        nst3 = NodeStyle()
        nst3["bgcolor"] = "lightblue"
        node.set_style(nst3)
#ts.mode = 'c'
#ts.arc_start = -90 # 0 degrees = 3 o'clock
#ts.arc_span = 90
ts.show_leaf_name = True
ts.show_branch_length = True
ts.show_branch_support = False
#ts.scale = 12 

ts.layout_fn = mylayout

anc.render("%%inline", tree_style=ts)
anc.render("../data/combined_tree/genome.png", tree_style=ts)
```

from ete3 import Tree, faces, AttrFace, TreeStyle, NodeStyle

def layout(node):
    if node.is_leaf():
        N = AttrFace("name", fsize=30)
        faces.add_face_to_node(N, node, 0, position="aligned")

def get_example_tree():

    # Set dashed blue lines in all leaves
    nst1 = NodeStyle()
    nst1["bgcolor"] = "LightSteelBlue"
    nst2 = NodeStyle()
    nst2["bgcolor"] = "Moccasin"
    nst3 = NodeStyle()
    nst3["bgcolor"] = "DarkSeaGreen"
    nst4 = NodeStyle()
    nst4["bgcolor"] = "Khaki"


    t = Tree("((((a1,a2),a3), ((b1,b2),(b3,b4))), ((c1,c2),c3));")
    for n in t.traverse():
        n.dist = 0

    n1 = t.get_common_ancestor("a1", "a2", "a3")
    n1.set_style(nst1)
    n2 = t.get_common_ancestor("b1", "b2", "b3", "b4")
    n2.set_style(nst2)
    n3 = t.get_common_ancestor("c1", "c2", "c3")
    n3.set_style(nst3)
    n4 = t.get_common_ancestor("b3", "b4")
    n4.set_style(nst4)
    ts = TreeStyle()
    ts.layout_fn = layout
    ts.show_leaf_name = False

    ts.mode = "c"
    ts.root_opening_factor = 1
    return t, ts

if __name__ == "__main__":
